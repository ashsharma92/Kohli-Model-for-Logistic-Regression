```python
import pandas as pd

filepath1='fixtures 2008.csv'
filepath2='fixtures 2009.csv'
filepath3='fixtures 2010.csv'
filepath4='fixtures 2011.csv'
filepath5='fixtures 2012.csv'
filepath6='fixtures 2013.csv'
filepath7='fixtures 2014.csv'
filepath8='fixtures 2015.csv'
filepath9='fixtures 2016.csv'
filepath10='fixtures 2017.csv'
filepath11='fixtures 2018.csv'
filepath12='fixtures 2019.csv'
filepath13='fixtures 2020.csv'
filepath14='fixtures 2021.csv'
filepath15='fixtures 2022.csv'
filepath16='fixtures 2023.csv'
filepath17='fixtures 2024.csv'
filepath18='fixtures 2025.csv'

df_2008= pd.read_csv(filepath1)
df_2009= pd.read_csv(filepath2)
df_2010= pd.read_csv(filepath3)
df_2011= pd.read_csv(filepath4)
df_2012= pd.read_csv(filepath5)
df_2013= pd.read_csv(filepath6)
df_2014= pd.read_csv(filepath7)
df_2015= pd.read_csv(filepath8)
df_2016= pd.read_csv(filepath9)
df_2017= pd.read_csv(filepath10)
df_2018= pd.read_csv(filepath11)
df_2019= pd.read_csv(filepath12)
df_2020= pd.read_csv(filepath13)
df_2021= pd.read_csv(filepath14)
df_2022= pd.read_csv(filepath15)
df_2023= pd.read_csv(filepath16)
df_2024= pd.read_csv(filepath17)
df_2025= pd.read_csv(filepath18)

```

```python
# Function to extract the team name
def extract_team_name(team_score_str):
    for i, char in enumerate(team_score_str):
        if char.isdigit():
            return team_score_str[:i].strip()
    return team_score_str.strip() # In case there are no digits

# Apply the function to the 'team_score' column to create a new 'team_name' column
df_2010['Team 1'] = df_2010['Team 1'].apply(extract_team_name)
df_2010['Team 2'] = df_2010['Team 2'].apply(extract_team_name)
df_2011['Team 1'] = df_2011['Team 1'].apply(extract_team_name)
df_2011['Team 2'] = df_2011['Team 2'].apply(extract_team_name)
df_2012['Team 1'] = df_2012['Team 1'].apply(extract_team_name)
df_2012['Team 2'] = df_2012['Team 2'].apply(extract_team_name)
df_2013['Team 1'] = df_2013['Team 1'].apply(extract_team_name)
df_2013['Team 2'] = df_2013['Team 2'].apply(extract_team_name)
df_2014['Team 1'] = df_2014['Team 1'].apply(extract_team_name)
df_2014['Team 2'] = df_2014['Team 2'].apply(extract_team_name)
df_2015['Team 1'] = df_2015['Team 1'].apply(extract_team_name)
df_2015['Team 2'] = df_2015['Team 2'].apply(extract_team_name)
df_2016['Team 1'] = df_2016['Team 1'].apply(extract_team_name)
df_2016['Team 2'] = df_2016['Team 2'].apply(extract_team_name)
df_2017['Team 1'] = df_2017['Team 1'].apply(extract_team_name)
df_2017['Team 2'] = df_2017['Team 2'].apply(extract_team_name)
df_2018['Team 1'] = df_2018['Team 1'].apply(extract_team_name)
df_2018['Team 2'] = df_2018['Team 2'].apply(extract_team_name)
df_2019['Team 1'] = df_2019['Team 1'].apply(extract_team_name)
df_2019['Team 2'] = df_2019['Team 2'].apply(extract_team_name)
df_2020['Team 1'] = df_2020['Team 1'].apply(extract_team_name)
df_2020['Team 2'] = df_2020['Team 2'].apply(extract_team_name)
df_2021['Team 1'] = df_2021['Team 1'].apply(extract_team_name)
df_2021['Team 2'] = df_2021['Team 2'].apply(extract_team_name)
df_2022['Team 1'] = df_2022['Team 1'].apply(extract_team_name)
df_2022['Team 2'] = df_2022['Team 2'].apply(extract_team_name)
df_2023['Team 1'] = df_2023['Team 1'].apply(extract_team_name)
df_2023['Team 2'] = df_2023['Team 2'].apply(extract_team_name)
df_2024['Team 1'] = df_2024['Team 1'].apply(extract_team_name)
df_2024['Team 2'] = df_2024['Team 2'].apply(extract_team_name)
df_2025['Team 1'] = df_2025['Team 1'].apply(extract_team_name)
df_2025['Team 2'] = df_2025['Team 2'].apply(extract_team_name)
```

```python
# List of DataFrames to concatenate
dataframes_to_union = [df_2008, df_2009,df_2010,df_2011,df_2012,df_2013,df_2014,df_2015,df_2016,df_2017,df_2018,df_2019,df_2020,df_2021,df_2022,df_2023,df_2024,df_2025]

# Perform the union (stacking on top)
df_union = pd.concat(dataframes_to_union)
```

```python
#reduce size of union dataset to include RCB matches only- need to extract team name only

RCB_data = df_union_cleaned[
    (df_union_cleaned['Team 1'].isin(['Royal Challengers Bangalore', 'Royal Challengers Bengaluru'])) |
    (df_union_cleaned['Team 2'].isin(['Royal Challengers Bangalore', 'Royal Challengers Bengaluru']))
]
```

```python
#join with kohli dataset

kohli_RCB_joined= pd.merge(left=kohli_stats, right=RCB_data, how='left', on='Match_ID')
```

```python
display(kohli_RCB_joined)
```

```python
final_data= kohli_RCB_joined[['SR','50','Result']]
```

```python
display(final_data)
```

```python
winning_teams = ['Royal Challengers Bangalore', 'Royal Challengers Bengaluru']
import numpy as np
# Create a new binary 'Win' column
final_data['Win'] = np.where(final_data['Result'].isin(winning_teams), 1, 0)

# Display the DataFrame with the new 'Win' column
display(final_data)

# You can now drop the original 'Result' column if you only need the binary outcome
final_data = final_data.drop('Result', axis=1)
print("\nDataFrame after dropping 'Result' column:")
display(final_data)
```

```python
display(final_data)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define features (X) and target (y)
X = final_data[['50']]
y = final_data['Win']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_proba = model.predict_proba(X_test)[:, 1]  # Probability of winning (class 1)
y_pred = model.predict(X_test)  # Binary predictions

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# You can also interpret the coefficient
display(f"\nCoefficient for Kohli_Half_Century: {model.coef_[0][0]}")
print(f"Intercept: {model.intercept_[0]}")
```
